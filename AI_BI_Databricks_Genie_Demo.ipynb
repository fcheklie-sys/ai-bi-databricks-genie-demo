```python
# Read the Superstore CSV file from ADLS Gen2 into a Spark DataFrame 
orders_df = spark.read.csv(
    f"abfss://{container}@{storage_account}.dfs.core.windows.net/Superstore.csv",
    header=True,
    inferSchema=True
)

display(orders_df.limit(10))

```

```python
# Standardize column names by trimming spaces, converting to lowercase, and replacing spaces with underscores
orders_df = orders_df.toDF(
    *[c.strip().lower().replace(" ", "_") for c in orders_df.columns]
)
display(orders_df.limit(10))
```

```python
%sql
-- List all available catalogs in the Databricks workspace (Unity Catalog discovery)
SHOW CATALOGS;
```

```python
%sql
-- List all schemas available in the current catalog
SHOW SCHEMAS;
```

```python
# Persist the raw ingested dataset as a Delta table in the Bronze layer for durable storage and downstream processing
orders_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("databricks_fikru_demo.default.superstore_bronze")
```

```python
%sql
-- Create the Silver-layer table by cleaning and standardizing data types from the Bronze layer
CREATE TABLE databricks_fikru_demo.default.superstore_silver AS
SELECT
  row_id,
  order_id,
  order_date,
  ship_date,
  ship_mode,
  customer_id,
  customer_name,
  segment,
  country,
  city,
  state,
  CAST(postal_code AS STRING) AS postal_code,
  region,
  product_id,
  category,
  sub_category,
  product_name,
  CAST(sales AS DOUBLE) AS sales,
  CAST(quantity AS INT) AS quantity,
  CAST(discount AS DOUBLE) AS discount,
  profit
FROM databricks_fikru_demo.default.superstore_bronze;
```

```python
%sql
-- Display column names, data types, and metadata for the Silver-layer Superstore table
DESCRIBE databricks_fikru_demo.default.superstore_silver;
```

```python
%sql
-- Check uniqueness at order + product level
SELECT order_id, product_id, COUNT(*) AS cnt
FROM databricks_fikru_demo.default.superstore_silver
GROUP BY order_id, product_id
HAVING cnt > 1;
```

```python
%sql
-- Create the Silver-layer table by normalizing the Bronze data to a clean orderproduct grain
-- Duplicate order_id + product_id rows are intentionally aggregated to preserve business value
CREATE OR REPLACE TABLE databricks_fikru_demo.default.superstore_silver AS
SELECT
  order_id,
  product_id,

  -- descriptive attributes (safe to MIN/MAX since identical)
  MIN(order_date)      AS order_date,
  MIN(ship_date)       AS ship_date,
  MIN(ship_mode)       AS ship_mode,
  MIN(customer_id)     AS customer_id,
  MIN(customer_name)   AS customer_name,
  MIN(segment)         AS segment,
  MIN(country)         AS country,
  MIN(city)            AS city,
  MIN(state)           AS state,
  MIN(postal_code)     AS postal_code,
  MIN(region)          AS region,
  MIN(category)        AS category,
  MIN(sub_category)    AS sub_category,
  MIN(product_name)    AS product_name,

  -- additive measures (must be summed)
  SUM(sales)           AS sales,
  SUM(quantity)        AS quantity,
  SUM(discount)        AS discount,
  SUM(profit)          AS profit

FROM databricks_fikru_demo.default.superstore_bronze
GROUP BY order_id, product_id;
```

```python
%sql
-- Create a Gold-layer view by enriching the Silver table with a derived profit flag for business analysis
CREATE OR REPLACE VIEW databricks_fikru_demo.default.superstore_gold_v AS
SELECT
  *,
  CASE
    WHEN profit > 0 THEN 'Profitable'
    ELSE 'Loss'
  END AS profit_flag
FROM databricks_fikru_demo.default.superstore_silver;
```

```python
%sql
-- Validate the Gold-layer profit flag by counting records in each profit category
SELECT
  profit_flag,
  COUNT(*) AS row_count
FROM databricks_fikru_demo.default.superstore_gold_v
GROUP BY profit_flag;
```

```python
%sql
-- Preview sample records from the Gold-layer view to verify business fields and profit classification
SELECT
  order_id,
  category,
  sub_category,
  sales,
  profit,
  profit_flag
FROM databricks_fikru_demo.default.superstore_gold_v
LIMIT 10;
```

```python
%sql
-- Add business-friendly metadata to Gold-layer columns to support BI tools and Genie semantic understanding
COMMENT ON COLUMN databricks_fikru_demo.default.superstore_gold_v.order_date
IS 'Date when the customer placed the order';

COMMENT ON COLUMN databricks_fikru_demo.default.superstore_gold_v.ship_date
IS 'Date when the order was shipped to the customer';

COMMENT ON COLUMN databricks_fikru_demo.default.superstore_gold_v.sales
IS 'Total sales amount for the order line item';

COMMENT ON COLUMN databricks_fikru_demo.default.superstore_gold_v.profit
IS 'Net profit for the order line item after costs';

COMMENT ON COLUMN databricks_fikru_demo.default.superstore_gold_v.quantity
IS 'Number of units sold in the order line item';

COMMENT ON COLUMN databricks_fikru_demo.default.superstore_gold_v.discount
IS 'Discount applied to the order line item as a percentage';

COMMENT ON COLUMN databricks_fikru_demo.default.superstore_gold_v.profit_flag
IS 'Indicates whether the order line item is profitable or resulted in a loss';
```
